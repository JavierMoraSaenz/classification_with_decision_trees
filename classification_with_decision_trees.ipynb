```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
# importamos librerias principales para trabajar
import os
os.chdir("C:/Users/ElMorado/OneDrive/Escritorio/LIBROS Y CURSOS/CURSO EBAC/Datasets")

df = pd.read_csv("drugs.csv")
df
#abrimos archivo con el que trabajaremos
```

```python
# asignamos los valores que tendran 'x' y 'y'
feature_calls = ['Age','Sex','BP','Cholesterol','Na_to_K']
x = df[feature_calls].values
y = df.Drug
```

```python
# Codificar las variables predictoras eliminando textos con preprocesamiento
from sklearn import preprocessing
#Recodificando  columna 'sex'
recod_sex = preprocessing.LabelEncoder()
recod_sex.fit(['F','M'])
x[:,1]= recod_sex.transform(x[:,1])
```

```python
#Recodificando  columna 'BP'
recod_BP = preprocessing.LabelEncoder()
recod_BP.fit(['HIGH','NORMAL','LOW'])
x[:,2] = recod_BP.transform(x[:,2])
```

```python
#Recodificando  columna de cholesterol
recod_chol = preprocessing.LabelEncoder()
recod_chol.fit(['HIGH','NORMAL','LOW'])
x[:,3] = recod_chol.transform(x[:,3])
```

```python
x[0:]
```

```python
len(y)
```

```python
# Dividision de grupos test y train
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.20,random_state=3)
```

```python
#Creacion de objeto de clasificacion del arbol de decision
clf = DecisionTreeClassifier() #por default gini

clf = clf.fit(X_train, y_train)

ypred= clf.predict(X_test)
```

```python
# Creacion de matriz de confusion
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,ypred)
cm
```

```python
# Reporte de clsificacion usando criterio gini
from sklearn.metrics import classification_report
print(classification_report(y_test,ypred))
```

```python
from sklearn import tree
import graphviz
import pydotplus
from PIL import Image

# Creacion de informacion par el arbol (reglas de decision)
dot_data = tree.export_graphviz(clf,out_file=None, feature_names = feature_calls, class_names = ['drugX','drugC','drugY','drugA','drugB'])

#Creacion de la grafica del arbol
graph = pydotplus.graph_from_dot_data(dot_data)
```

```python
# exportamos imagen del arbol de decision gini
graph.write_png("Drugsgini.png")
image = Image.open("Drugsgini.png")
image.show()
```

```python
# Creamos la clasificacion con el criterio entropia
clf = DecisionTreeClassifier(criterion='entropy') 

clf = clf.fit(X_train, y_train)

ypred= clf.predict(X_test)
```

```python
# Creacion de matriz de confusion
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,ypred)
cm
```

```python
# Reporte de clsificacion usando criterio entropy
from sklearn.metrics import classification_report
print(classification_report(y_test,ypred))
```
